```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
model = KMeans(n_clusters=2)
```

```python
data = pd.read_csv(r"D:\Data\AI & ML\bupa.csv")
data = data[['mcv', 'alkphos']]
```

```python
dimension = TSNE(n_components=2, perplexity=40)
```

```python
new_data = dimension.fit_transform(data)
```

```python
y = model.fit_predict(new_data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(new_data[:, 0], new_data[:, 1], c=y)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=90, c='red')
plt.show()
```

```python
import mplcursors
```

```python
data = pd.read_csv(r"D:\Data\AI & ML\bupa.csv")
data2 = data[['mcv', 'sgpt', 'sgot']]
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data2)
```

```python
data2['cluster'] = kmeans.labels_
```

```python
plt.scatter(data2['mcv'], data2['sgpt'], c=data2['cluster'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=90)

mplcursors.cursor(hover=True).connect("add", lambda sel: sel.annotation.set_text(
    f"mcv: {data2['mcv'][sel.target.index]}, sgpt: {data2['sgpt'][sel.target.index]}"
))

plt.show()
```
